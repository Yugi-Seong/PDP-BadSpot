### 양품 = 0, 명점 = 1

```python
for i in range(4000):
    if df['JUDGE'][i] == '양품' :
        df['JUDGE'][i] = 0
    else :
        df['JUDGE'][i] = 1
```

```python
df[df['JUDGE']==0].count()
# df.groupby("JUDGE").count
```

### LOT-101 => 101

```python
for i in range(4000):
    df['LOT_ID'][i] = int(df['LOT_ID'][i][4:])
    
df.head(3)
```

### PNL-01 => 01

```python
for i in range(4000):
    df['PANEL_ID'][i] = int(df['PANEL_ID'][i][4:])
    
df.head(3)
```
