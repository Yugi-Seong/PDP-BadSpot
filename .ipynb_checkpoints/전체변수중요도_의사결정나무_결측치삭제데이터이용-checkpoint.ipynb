```python
df = pd.read_csv("data/결측치삭제.csv", engine="python",encoding="cp949")
df.head()
```

```python
df.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier as DTC
```

```python
df_y=df[["JUDGE"]]
df_y
```

```python
df_x = df.drop("JUDGE", axis=1, inplace=False)
df_x
```

```python
df_x.info()
```

모두 float64 타입 = > 더미변수 불필요 

```python
#최종모델 
tree_final = DTC(random_state = 1234, max_depth = 4)
tree_final.fit(df_x,df_y)
```

```python
# 설명변수 중요도 확인 및 테이블로 저장
v_feature_name = df_x.columns
```

```python
df_importance = pd.DataFrame()
df_importance['Feature'] = v_feature_name
df_importance['Importance'] = tree_final.feature_importances_

# 중요도 순으로 정렬
df_importance.sort_values('Importance', ascending=False, inplace=True)
df_importance.round(3)
```
